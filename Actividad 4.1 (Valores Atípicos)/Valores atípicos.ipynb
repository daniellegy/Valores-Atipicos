**ACTIVIDAD 4.1 (VALORES ATÍPICOS)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
data
```

```python
data.info()
```

```python
cuantitativas = data.iloc [:,7:14] #1:23
cualitativas=data.iloc[:,0:7] #[0]
```

```python
cualitativas
```

```python
cuantitativas.info()
```

```python
fig= plt.figure (figsize=(15,8))
cuantitativas.plot(kind="box", vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
Promedio= y.mean()
print("Limite superior permitido\n",Limite_Superior)
print("Promedio\n", Promedio)
print("Limite inferior permitido\n", Limite_Inferior)

```

```python
valores_nulos=cuantitativas.isnull().sum()
valores_nulos
```

```python
dataDE=cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
dataDE
```

```python
valores_nulos=dataDE.isnull().sum()
valores_nulos
```

```python
data_clean_DE=dataDE.copy()
data_clean_DE=data_clean_DE.fillna(round(dataDE.mean(),1))
data_clean_DE
```

```python
valores_nulos_DE=data_clean_DE.isnull().sum()
valores_nulos_DE
```

```python
fig= plt.figure (figsize=(15,8))
data_clean_DE.plot(kind="box", vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

**PROCEDIMIENTO CUANTILES PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
x=cuantitativas

percentil25=x.quantile(0.25) #Q1
percentil75=x.quantile(0.75) #Q3
iqr= percentil75 - percentil25

Limite_Superior_iqr=percentil75 + 1.5*iqr
Limite_Inferior_iqr=percentil25 - 1.5*iqr
print("Limite superior permitido\n", Limite_Superior_iqr)
print("Limite inferior permitido\n", Limite_Inferior_iqr)
```

```python
valores_nulos_iqr=cuantitativas.isnull().sum()
valores_nulos_iqr
```

```python
data_iqr=cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_iqr
```

```python
valores_nulos_iqr=data_iqr.isnull().sum()
valores_nulos_iqr
```

```python
data_clean_iqr=data_iqr.copy()
data_clean_iqr=data_clean_iqr.fillna(round(data_iqr.mean(),1))
data_clean_iqr
```

```python
valores_nulos_iqr_clean=data_clean_iqr.isnull().sum()
valores_nulos_iqr_clean
```

**Verificamos ambos tipos de cambio para valores atípicos**

```python
fig= plt.figure (figsize=(15,8))
data_clean_DE.plot(kind="box", vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

```python
fig= plt.figure (figsize=(15,8))
data_clean_iqr.plot(kind="box", vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```
